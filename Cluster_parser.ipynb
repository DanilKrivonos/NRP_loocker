```python
import json, os, sys
```

```python
#Change your path
paths = os.listdir('/Users/danilkr/Downloads/mibig_json_2.0/')
```

```python
save_file = open('/Users/danilkr/Desktop/pk_C.txt', 'w')
save_file.write('MiBig\tBiosin_class\tGen\tAdenilation\tCondencetion\n')
for path in paths:
    with open('/Users/danilkr/Downloads/mibig_json_2.0/' + path, encoding='utf-8-sig') as js:
        try:
            j = json.load(js)
            if 'NRP' in j['cluster']['biosyn_class']:     
           
                if 'nrp' not in j['cluster']:
                    continue
                if 'nrps_genes' not in j['cluster']['nrp']:
                    continue
                print('Checking {}...'.format(path))
                for mod in j['cluster']['nrp']['nrps_genes']:
                    Adenilation = []
                    Condensation = []
                    for dom in mod['modules']:                        
                        if 'a_substr_spec' in dom:
                            Adenilation.append(dom['a_substr_spec'])
                        elif 'c_dom_subtype' in dom:
                            Condensation.append(dom['c_dom_subtype'])

                    if Adenilation == []:
                        Adenilation = 'NaN'
                    if Condensation == []:
                        Condensation = 'NaN'
                    save_file.write('{MiBig}\t{Biosin_class}\t{gen}\t{Adenilation}\t{Condensation}\n'.format(MiBig = str(j['cluster']['mibig_accession']),
                                                                                                      Biosin_class = str(j['cluster']['biosyn_class']), 
                                                                                                      gen = mod['gene_id'],
                                                                                                      Adenilation = str(Adenilation), 
                                                                                                      Condensation = str(Condensation)
                                                                                                      ))
                print('Done!')
            else:
                continue
        except:
            print('I can not parse this {}:('.format(path))
            
```
