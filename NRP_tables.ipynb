```python
import sys, os
from Bio.SeqIO import parse
from pyteomics.fasta import read
```

```python
faa = read('/Users/danilkr/Desktop/only_my.faa')

gbk_path = os.listdir('/Users/danilkr/Downloads/mibig_gbk_2.0/')
```

```python
sawe_file = open('/Users/danilkr/Desktop/Domens.txt', 'w')
sawe_file.write('ID\tProt_id\tCoordinate_in_gen\tDomen_type\tCoordinate_in_pept\tNote\tTranslate\n')
for gen in faa:
    for gbk in gbk_path:
        if gen.description.split('|')[0] in gbk:
            gbk_s = parse('/Users/danilkr/Downloads/mibig_gbk_2.0/' + gbk, 'genbank')
            for line in gbk_s:
                for i in line.features:
                    if i.qualifiers.get('NRPS_PKS') != None:
                        if 'NRP' in i.qualifiers.get('NRPS_PKS')[-1]:
                            print('Looking at {}'.format(i.qualifiers.get('protein_id')))
                            for domen in i.qualifiers.get('NRPS_PKS'):
                                if 'Domain' in domen:
                                    #print(domen)
                                    l = int(domen.split()[2].split('-')[0][1: ])
                                    r = int(domen.split()[2].split('-')[1][: -2])
                                    Domen = domen.split()[1]
                                    coord_pept = '[{}]'.format(domen.split()[2][1: -2])
                                    translate = i.qualifiers.get('translation')[0][l: r]

                                #else:
                                    #Domen = domen
                                    #coord_pept = '?'
                                    #translate = '?'


                            sawe_file.write('{ID}\t{Prot_id}\t{Coord_in_genome}\t{Dom}\t{Cord_p}\t{Note}\t{Transl}\n'.format(ID = line.id,
                                                                                                                         Prot_id = i.qualifiers.get('protein_id'),
                                                                                                                         Coord_in_genome = i.location,
                                                                                                                         Dom = Domen,
                                                                                                                         Cord_p = coord_pept,
                                                                                                                         Note = i.qualifiers.get('NRPS_PKS')[-1][5: ],
                                                                                                                         Transl = translate
                                                                                                                         ))
                        
                    
                    
sawe_file.close()           
print('Done!')
```
